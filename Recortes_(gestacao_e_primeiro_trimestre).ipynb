```python
dataset = pd.read_csv(path+arquivo, sep=',')
dataset
```

```python
dataset = dataset.loc[dataset['CLASSI_FIN'] == 1, dataset.columns]
dataset
```

```python
dataset.NU_IDADE_N.describe()
```

```python
def distribuicao_idade(data):
  print("describe de pessoas que tiveram zika em hora")
  print(data.loc[data['NU_IDADE_N'] < 2000, 'NU_IDADE_N'].describe())
  print("describe de pessoas que tiveram zika em dias")
  print(data.loc[(data['NU_IDADE_N'] > 2000) & (data['NU_IDADE_N'] < 3000), 'NU_IDADE_N'].describe())
  print("describe de pessoas que tiveram zika em mes")
  print(data.loc[(data['NU_IDADE_N'] > 3000) & (data['NU_IDADE_N'] < 4000), 'NU_IDADE_N'].describe())
  print("describe de pessoas que tiveram zika em ano")
  print(data.loc[data['NU_IDADE_N'] > 4000, 'NU_IDADE_N'].describe())
```

```python
distribuicao_idade(dataset)
```
